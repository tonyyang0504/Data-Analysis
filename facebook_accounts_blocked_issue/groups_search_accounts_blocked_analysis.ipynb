#### 编写一个函数计算每个被封账号进行搜群任务的开始时间和结束时间，并计算进行搜群任务所用时间（以小时为单位）。

```python
def caculate_action_time(server_logs, server_names):
    action_time = server_logs.groupby('Account').agg({'Date': ['max', 'min']}).Date.loc[server_names]
    action_time['end_time'] = pd.to_datetime(action_time['max'])
    action_time['start_time'] = pd.to_datetime(action_time['min'])
    action_time['diff_time'] = action_time['max'] - action_time['min']
    action_time['diff_hours'] = action_time['diff_time'].dt.total_seconds() / 3600
    
    return action_time[['start_time', 'end_time', 'diff_time', 'diff_hours']]
```

#### 编写一个函数分别计算5台机器人被封账号执行搜群任务所用时间和完成数量的平均值和标准差。

```python
def caculate_average_and_std(server_normalized_data):
    server_action_time_median = round(server_normalized_data['action time'].median())
    server_action_time_average = round(server_normalized_data['action time'].mean())
    server_action_time_std = round(server_normalized_data['action time'].std())
    server_action_count_median = round(server_normalized_data['action count'].median())
    server_action_count_average = round(server_normalized_data['action count'].mean())
    server_action_count_std = round(server_normalized_data['action count'].std())
    server_account_amount = len(server_normalized_data['action time'])
    result = f'******************************************\n      server action time median--> {server_action_time_median}\n      server action time average--> {server_action_time_average}\n      server action time std--> {server_action_time_std}\n      server action count_median--> {server_action_count_median}\n      server action count_average--> {server_action_count_average}\n      server action count std--> {server_action_count_std}\n      server account amount--> {server_account_amount}\n******************************************\n'

    return print(result)
```

```python
server_normalized_data_dict = {'server1_statistics_results': server1_normalized_data, 
                               'server155_statistics_results': server155_normalized_data,
                               'server183_statistics_results': server183_normalized_data,
                               'server183_old_statistics_results': server183_old_normalized_data,
                               'server194_5_statistics_results': server194_5_normalized_data,
                               'server194_6_statistics_results': server194_6_normalized_data,
                               'server27_statistics_results': server27_normalized_data}

for i, j in server_normalized_data_dict.items():
    print(f'      {i}      ')
    caculate_average_and_std(j)
```
