```python
import pandas as pd
from xlrd import xldate_as_datetime
```

```python
df = pd.read_excel("Team outside's Report - Copy.xlsx")
```

```python
df
```

```python
def convert_date_format(date):
    for i in range(date.shape[0]):
        try:
            date.at[i] = xldate_as_datetime(date.at[i], 0).strftime('%Y-%m-%d')
        except Exception as e:
            continue
```

```python
convert_date_format(df.Date)
```

```python
set(df.Date)
```

```python
df.Date.mask(df.Date.str.contains(','), df.Date.str.split()[0], inplace=True)
df['Date'] = pd.to_datetime(df.Date).astype('str')
```

```python
set(df.Date)
```

```python
df['Date'].where(df.Date.str[:4] == '2022', '2022' + df.Date.str[4:], inplace=True)
```

```python
condition1 = df.Date.str.slice(5, 7).astype('int') > 9
condition2 = df.Date.str.slice(5, 7).astype('int') < 7
condition3 = (df.Date.str.slice(5, 7).astype('int') == 7) & (df.Date.str.slice(8, ).astype('int') < 12)

replaced_value = '2022' + '-' + df.Date.str[8:] + '-' + df.Date.str[5:7]

df['Date'].mask(condition1, replaced_value, inplace=True)
df['Date'].mask(condition2, replaced_value, inplace=True)
df['Date'].mask(condition3, replaced_value, inplace=True)
```

```python
result = df.pivot_table(index='Date', columns='Name', values='#Name', aggfunc='count').fillna(0)
result
```
